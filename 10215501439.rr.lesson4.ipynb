```python
"""请编写一个函数，
    利用 辗转除2取余法（不要使用int的直接转换） 
    实现将我们常见的十进制IPv4地址例如203.179.25.37转换为32位长的二进制地址。
"""
#成功！
def ten22(n):#辗转除2取余法   
       #利用的基本对象：列表（列表可以存放一切类型，类似于一个类）
        b = []  # 存储余数
        while True:  # 一直循环，商为0时利用break退出循环
            s = n // 2  # 商
            y = n % 2  # 余数
            b = b + [y]  # 每一个余数存储到b中（加列表简单粗暴）
            if s == 0:
                break  # 余数为0时结束循环
            n = s
        b.reverse()  # 使b中的元素反向排列
        res=int(''.join(list(map(str,b))))#将列表中的数字（可能是数字字符串，可能是单个数字字符）转化为int型数字
        return res
    
def convert_ip_address_to_dec(address):
    firstlist=[]
    for i in range(4):#注意，i取到0,1,2,3
        firstlist.append(address.split('.',3)[i])#利用split函数，和显著的分隔符‘。’，3表示一共分为3+1段，【i】表示选取第i段
    secondlist=[]
    for add in firstlist:
        secondlist.append(ten22(int(''.join(list(map(str,add))))))#传入ten22函数的是数字，故转换一下
    result=str(secondlist[0])+'.'+str(secondlist[1])+'.'+str(secondlist[2])+'.'+str(secondlist[3])
    return result

address='203.179.25.37'
print(convert_ip_address_to_dec(address))
#难题，类型转换
```

```python
"""
请创建一个类模拟实现数据结构中的“队列”，
类中应包含入队、出队、取队首/队尾元素等方法，并进行测试。
"""
class Quene():
    def _init_(self):
        self._items = []
    def size(self):
        return len(self._items)
    def isempty(self):
        if len(self._items)==0:
            return True
        else:
            return False
    def inquene(self,element):
        self._items.append(element)
    def pop(self):
        try:
            return self._items.pop(0)
        except: 
            print('ERROR:Quene is empty now!!')
    def peek(self):
        try:
            return self._items[0]
        except:
            print('ERROR:Quene is empty now!!')

q=Quene()
#inquene(2)
print(q.pop()) # 堆栈目前为 [ ]
print(q.pop()) # 错误，堆栈为空

print('**********')

q.inquene(3.5)  # 堆栈目前为[3.5]
q.inquene(2.7)  # 堆栈目前为[3.5, 2.7]
print(q.peek())
print(q.size())
print(q.isempty())
```
